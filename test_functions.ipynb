```python
import datetime, re

# Inicializa una lista para almacenar los años de lanzamiento
releases_year = []

# Itera a través de las fechas en la columna 'release_date' de un DataFrame df_games
for date in df_games['release_date']:
    try:
        # Intenta convertir la fecha al año utilizando el formato "%Y-%m-%d"
        releases_year.append(datetime.datetime.strptime(date, "%Y-%m-%d").year)
    
    except (ValueError, TypeError):
        # Si hay un error al convertir la fecha, se maneja aquí
        if date is not None:
            # Extrae los dígitos de la fecha utilizando una expresión regular
            edited_date = re.findall(r'\d+', date)
            
            # Verifica si los dígitos extraídos forman una lista
            if type(edited_date) is list:
                # Comprueba si la lista de dígitos no está vacía
                if len(edited_date) > 0:
                    # Inicializa una bandera para indicar si se ha encontrado un año válido
                    sum_date = False
                    # Inicializa un contador para llevar un registro de los números válidos encontrados
                    cont = 0
                    for num in edited_date:
                        # Comprueba si el número está en el rango de años esperado (1985-2025)
                        if 1985 < int(num) < 2025:
                            # Agrega el año a la lista releases_year y aumenta el contador
                            releases_year.append(int(num))
                            cont += 1
                            sum_date = True
                            # Sale del bucle una vez que se ha encontrado un año válido
                            break
                
                    # Si no se encontró un año válido, agrega NaN a la lista
                    if sum_date is False:
                        releases_year.append(np.nan)
                        
                    # Si el contador es mayor que 1, imprime la fecha (para depuración)
                    if cont > 1:
                        print(date)
                else:
                    # Si la lista de dígitos está vacía, agrega NaN a la lista
                    releases_year.append(np.nan)
            else:
                # Si no es una lista, imprime la lista de dígitos (para depuración)
                print(edited_date)
        else:
            # Si la fecha es None, agrega NaN a la lista
            releases_year.append(np.nan)

# Imprime la cantidad de fechas en df_games['release_date'] y la cantidad de elementos en releases_year
print(f"Cantidad de fechas en df_games['release_date']: {len(df_games['release_date'])}. Cantidad de elementos en released_year: {len(releases_year)}")
```

```python
lista_fechas_reviews = []

for posted in merge_UserRecommend['posted']:
    import re

    # Define una expresión regular para buscar el año en el texto
    patron = r'\b\d{4}\b'  # Coincide con una secuencia de 4 dígitos

    # Utiliza re.search() para encontrar la primera coincidencia del patrón en el texto
    match = re.search(patron, posted)

    if match:
        # Si se encontró una coincidencia, obtén el año como un entero
        año = int(match.group())
        lista_fechas_reviews.append(int(año))
    else:
        print(f"{posted}")
        lista_fechas_reviews.append(2016)


merge_UserRecommend['posted_year'] = lista_fechas_reviews
```

```python
def tiempo_por_juego(genero:str):
    row = df_PlayTimeGenre[df_PlayTimeGenre['genero']==genero].T
    list_genres = list(df_PlayTimeGenre['genero'].values)

    max_anio = 0
    for año, item in row.iterrows():
        try:    
            if item.values[0] > max_anio:
                max_anio = año
                max_playtime = item.values[0]
            return {
                f"Año de lanzamiento con más horas para género {genero}": max_anio,
                f"Cantidad de horas jugadas": max_playtime
            }
        except TypeError:
            pass
        except IndexError:
            return None
```

```python
df_UserRecommend = pd.read_json("./data/data_to_functions/UserRecommend.json")

def UserRecommend(year:int):
    if not isinstance(year, int):
        raise TypeError("The function wait a integer.")

    try: 
        row = list(df_UserRecommend[year].values)

        return [
            {"Puesto 1": row[0][0]},
            {"Puesto 2": row[1][0]},
            {"Puesto 3": row[2][0]}
        ]
    except TypeError:
        return None
    except KeyError:
        return {"Error": f"The function doesn't find {year}'s recommendations."}
```

```python
df_UserNotRecommend = pd.read_json("./data/data_to_functions/UserNotRecommend.json")

def UserNotRecommend(year:int):
    if not isinstance(year, int):
        raise TypeError("The function wait a integer.")

    try:
        row = list(df_UserNotRecommend[year].values)

        not_recommend = {}
        for ix, no_game in enumerate(row):
            try:
                not_recommend[f"Puesto {ix + 1}"] = no_game[0]
            except TypeError:
                continue

        return not_recommend
    except KeyError:
        return {"Error": f"The function doesn't find {year}'s recommendations."}
```

```python
def sentiment_analysis(year:int):
    if not isinstance(year, int):
        raise TypeError("The function wait a integer.")            

    try:
        year = str(year)
        sentiments = list(df_sentiment_analyisis['Sentiment'].values)
        values = list(df_sentiment_analyisis[f"{year}"])

        dicc_sentiments = {}
        for index in range(0,3):
            dicc_sentiments[sentiments[index]] = values[index]
        return dicc_sentiments

    except IndexError:
        return {"Error": f"The function doesn't find {year}'s recommendations."}
```
